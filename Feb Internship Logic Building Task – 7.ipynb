# 1.Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(f"{self.name} is {self.status}")
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

# 2.Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print(f"Employee Name: {self.name}")
        print(f"Employee ID: {self.emp_id}")
        print(f"Department: {self.department}")
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

# 3.Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def save_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
contact = Contact("Anita", "9876543210")
contact.save_contact()
```

# 4.Product Price Tag Generator


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def display_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: {self.price}")
item = Product("Headphones", 2499)
item.display_price_tag()
```

# 5.Movie Rating Display System

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating
    def show_details(self):
        print(f"Movie Name: {self.title}")
        print(f"Rating: {self.rating}")
film = Movie("Inception", 4.8)
film.show_details()
```

# 6.Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")
order = Delivery("Suman", "Hyderabad")
order.print_details()
```
