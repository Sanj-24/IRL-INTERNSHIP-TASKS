2.Pass/Fail Analyzer

```python
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for i, score in enumerate(marks,start=1):
    if score>=50:
        print(f"Student {i}:{score}->Pass")
        pass_count+=1
    else:
        print(f"Student {i}:{score}->Fail")
        fail_count+=1
print("Final Result")
print("Total Students:",len(marks))
print("Passed:",pass_count)
print("Failed:",fail_count)
```

3.Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    cleaned=name.strip().lower()
    cleaned_names.append(cleaned)
print("Cleaned Names:",cleaned_names)
```

4.Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length=len(msg)
    print(f"Message:{msg} Length:{length}")
    if length>10:
       print("Note:This message is longer than 10 characters")
```

5.Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0
for entry in logs:
    if entry =="ERROR":
        error_count+=1
print("Total ERROR Entries:",error_count)
```
