# 1.Mobile Recharge Validation System

```python
def recharge_validation():
    allowed_plans = (199, 299, 399, 599)
    while True:
        recharge = int(input("Enter recharge amount:"))
        if recharge >= 50:
            if recharge in allowed_plans:
                print("Recharge completed successfully")
                break
            else:
                print("Plan not available")
        else:
            print("Minimum recharge is 50")
recharge_validation()
```

# 2.Inventory Reorder Alert System

```python
def reorder_check(products):
    for item in products:
        if products[item] < 15:
            print(item, "Reorder Alert")
        else:
            print(item, "Stock OK")
store_stock = {
    "Soap": 8,
    "Shampoo": 18,
    "Rice": 12,
    "Flour": 30
}
reorder_check(store_stock)
```

# 3.Student Result Processing System

```python
def calculate_result(mark_list):
    total_marks = 0
    for mark in mark_list:
        total_marks += mark
    average = total_marks / len(mark_list)
    if average >= 50:
        return "Pass"
    return "Fail"
student_marks = [45, 60, 55, 50]
print("Final Result:", calculate_result(student_marks))
```

# 4.Cab Fare Estimator with Retry Option

```python
def cab_fare(distance, peak):
    total_fare = 50 + (distance * 12)
    if peak:
        total_fare += total_fare * 0.25
    return total_fare
while True:
    km = float(input("Enter travel distance (km): "))
    peak_time = input("Peak hour? (yes/no): ").lower() == "yes"
    print("Total Fare: ", cab_fare(km, peak_time))
    repeat = input("Calculate again? (yes/no): ").lower()
    if repeat != "yes":
        break
```

# 5.Employee Attendance Eligibility Checker

```python
def eligibility_check(attendance):
    present_count = 0
    for status in attendance:
        if status == "P":
            present_count += 1
    attendance_percentage = (present_count / len(attendance)) * 100
    if attendance_percentage >= 75:
        return "Eligible"
    return "Not Eligible"
attendance_data = ["P", "A", "P", "P", "P", "A", "P", "P"]
print("Eligibility Status:", eligibility_check(attendance_data))
```

# 6.Password Strength Checker

```python
def password_validator(password):
    digit_present = False
    special_present = False
    if len(password) < 8:
        return "Weak Password"
    for char in password:
        if char.isdigit():
            digit_present = True
        elif char in "@#$":
            special_present = True
    if digit_present and special_present:
        return "Strong Password"
    return "Weak Password"
user_password = input("Create password: ")
print(password_validator(user_password))
```
