# 1.Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

top_performers = []
highest_score = max(employees.values())

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print(f"Top performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

# 2.Search Query Keyword Analysis

```python
import string
query = "Buy mobile phone buy phone online"
query = query.lower()
query = query.translate(str.maketrans("", "", string.punctuation))
words = query.split()
word_counts = {}
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1
repeated_keywords = {}
for word, count in word_counts.items():
    if count > 1:
        repeated_keywords[word] = count
print(repeated_keywords)
```

#3.Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# 4.Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_counts = {}
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_counts:
        domain_counts[domain] += 1
    else:
        domain_counts[domain] = 1
total_users = len(emails)
domain_percentages = {}
for domain, count in domain_counts.items():
    percentage = (count / total_users) * 100
    domain_percentages[domain] = round(percentage)
for domain, percentage in domain_percentages.items():
    print(f"{domain}: {percentage}%")
```

#5.Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
spike_days = []
for day, value in enumerate(sales, start=1):
    if value > average_sales * 1.3:
        spike_days.append((day, value))
print("Days with Sales Spikes (Day, Value):")
for day, value in spike_days:
    print(f"Day {day}: {value}")
```

# 6.Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_counts = {}
for uid in user_ids:
    if uid in id_counts:
        id_counts[uid] += 1
    else:
        id_counts[uid] = 1
print("Duplicate User IDs:")
for uid, count in id_counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
