# 1.Smart Parking Lot Management System

```python
def manage_parking(capacity,activity_log):
    current_count=0
    for action in activity_log:
        if action=="IN":current_count+=1
        elif action=="OUT" and current_count>0:current_count-=1
    lot_status="Available" if current_count<capacity else "Full"
    print("Currently Parked Vehicles:",current_count)
    print("Parking Status:",lot_status)
manage_parking(50,["IN","IN","IN","OUT","IN","IN","OUT"])
```

# 2.Online Food Delivery Time Estimator

```python
def estimate_delivery(distance_km,traffic_level,weather_condition):
    time_minutes=distance_km*3
    if traffic_level=="High":time_minutes+=20
    elif traffic_level=="Medium":time_minutes+=10
    if weather_condition=="Rainy":time_minutes+=11
    elif weather_condition=="Stormy":time_minutes+=30
    print("Estimated Delivery Time:",time_minutes,"minutes")
estimate_delivery(8,"High","Rainy")
```

# 3.Movie Theatre Seat Occupancy Analyzer

```python
def analyse_seats(total,booked_list):
    booked_count=len(booked_list)
    percentage=int((booked_count/total)*100)
    if percentage==100:status="Housefull"
    elif percentage>=75:status="Almost Full"
    else:status="Available"
    print("Occupancy:",str(percentage)+"%")
    print("Show Status:",status)
analyse_seats(200,[1]*150)
```

# 4.Cloud Server Load Classification System

```python
def classify_server(cpu_data):
    avg_usage=int(sum(cpu_data)/len(cpu_data))
    if avg_usage<50:category="Normal"
    elif avg_usage<=80:category="Warning"
    else:category="Critical"
    print("Average CPU Load:",str(avg_usage)+"%")
    print("Server Status:",category)
classify_server([45,60,70,85,90])
```

# 5.Smart Classroom Resource Usage Monitor

```python
def resource_monitor(usage_hours):
    flagged=[item for item,hours in usage_hours.items() if hours>8]
    alert="Yes" if flagged else "No"
    print("Overused Resources:",", ".join(flagged) if flagged else "None")
    print("Energy Alert:",alert)
resource_monitor({"Projector":6,"AC":9,"Lights":4})
```

# 6.Online Event Registration Capacity Controller

```python
def registration_manager(limit,total_registrations):
    confirmed=min(limit,total_registrations)
    waitlisted=max(0,total_registrations-limit)
    status="Closed" if total_registrations>=limit else "Open"
    print("Confirmed Registrations:",confirmed)
    print("Waitlisted Users:",waitlisted)
    print("Registration Status:",status)
registration_manager(100,105)
```
